```python
#Task 1
class Book ():


    def __init__(self,title:str, author:str,isbn:str):

        if not isinstance(title,str):
            raise TypeError("Book title must be a string!")
        self.title = title
        if not isinstance(author,str):
            raise TypeError("Author must be a string!")
        self.author = author
        if not isinstance(isbn,str):
            raise TypeError("ISBN must be a string!")
        self.isbn = isbn
        self.is_checked_out = False

     #defining Methods


    def display_info(self):
        print( f"{self.title}  is the tile of the book it is written by {self.author} and its isbn is {self.isbn}" )
        if self.is_checked_out == False:
            print(f"{self.title} is available")
        else:
            print(f"{self.title} is not available")
    def check_out(self):
        if self.is_checked_out == False:
            print(f"{self.title} has been checked out")
        else:
              print(f"{self.title} is already checked out")
    def check_in(self):
        if self.is_checked_out == True:
            print(f"{self.title} has been checked in")
        else:
            print(f"{self.title} is already available")

    #Task 2
class Library(Book):
    def __init__(self,name):
        self.name = name
        self.books = []
    def add_book(self,book_object):
        self.books.append(book_object)
        print(f"{book_object.title} added to {self.name}")
    def find_book_by_title(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book
        print(f"Book with title '{title}' not found.")
        return None
    def check_out_book(self,title):
        book = self.find_book_by_title(title)
        if book:
            book.check_out()

    def check_in_book(self,title):
        book = self.find_book_by_title(title)
        if book:
            book.check_in()

    # list_available_books(self):
    def list_available_books(self):
        mybook = [book for book in self.books if not book.is_checked_out]
        if not mybook:
            print("No books are available")
        else:
            for book in self.books:
                book.display_info()
                print(f"All books in {self.name}:")

    def list_all_books(self):
            print(f"All books in {self.name}:")
            for book in self.books:
                book.display_info()

#task 3 testing

my_lab = Library("City Central Library")
book_1 = Book("The Thousands Splendid Suns", "Khalid Hussaini","KHA0321")
book_2 = Book("Ikigai","Hector Gracia", "IK0321")
book_3 = Book("The Lesson of History"," Ariel Durant and Will Durant","Less0321")
my_lab.add_book(book_1)
my_lab.add_book(book_2)
my_lab.add_book(book_3)
my_lab.list_all_books()
my_lab.list_available_books()
my_lab.check_out_book("Ikigai")
my_lab.check_out_book("Ikigai")
my_lab.list_all_books()
my_lab.check_in_book("Ikigai")
my_lab.check_in_book("The Thousands Splendid Suns")
my_lab.list_all_books()
my_lab.find_book_by_title("Ikigai")
my_lab.find_book_by_title("The Subtle Art of Not Giving a F*ck")
```
